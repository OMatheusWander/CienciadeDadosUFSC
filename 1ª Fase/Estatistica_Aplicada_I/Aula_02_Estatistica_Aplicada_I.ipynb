```r
set.seed(2026)

runif(1)
```

```r
genero = sample(c("Feminino", "Masculino"), size = 10000, replace= T, prob = c(0.2, 0.8))

table(genero)
```

```r
sangue = sample(c("A", "B", "AB", "O"), size = 10000, replace= T, prob = c(0.4, 0.11, 0.04, 0.45))

table(sangue)
```

```r
residencia = sample(paste("Residência",1: 2500),
             size= 10000,
             replace = T)

max(table(residencia))
min(table(residencia))

table(table(residencia))
```

```r
renda = numeric(1000)

renda[genero == "Masculino"] = rexp(sum(genero == "Masculino"), rate = 1/4000)

renda[genero == "Feminino"] = rexp(sum(genero == "Feminino"), rate = 1/4000)


dados = data.frame(genero, sangue, residencia, renda)

str(dados)
dim(dados)
head(dados)
tail(dados)
```

```r
#Amostra Aleatória Simples

library(dplyr)
sample_n(dados,10)

#Amostra Estratificada

?slice_sample(dados,)


slice_sample( dados(genero == "Masculino"))
```

```r
dados |>
  filter(sangue == "A") |>
  slice_sample(n = 5 ) # Seleciona 1 linha aleatória (exemplo)
```

```r
dadosm = dados[genero == "Masculino",]
dadosf = dados[genero == "Feminino",]

sample_n(dadosm,8)
sample_n(dadosf,2)
```

```r
#Amostra por Conglomerado

res_sorteadas = sample(unique(dados$residencia), 10)
res_sorteadas
```

```r
dados[dados$residencia %in% res_sorteadas,]
```
