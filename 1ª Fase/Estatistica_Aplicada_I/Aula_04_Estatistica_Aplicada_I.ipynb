```r
# Criando os vetores com os dados de cada coluna
estudante <- 1:20

genero <- c("F", "M", "F", "M", "F", "M", "F", "F", "M", "F",
            "M", "F", "M", "F", "M", "F", "M", "F", "M", "F")

turno <- c("Noturno", "Noturno", "Matutino", "Noturno", "Vespertino",
           "Noturno", "Matutino", "Noturno", "Vespertino", "Matutino",
           "Noturno", "Vespertino", "Matutino", "Noturno", "Vespertino",
           "Matutino", "Noturno", "Vespertino", "Matutino", "Noturno")

n_disciplinas <- c(5, 6, 4, 5, 3, 6, 4, 5, 3, 4,
                   6, 4, 5, 6, 4, 3, 5, 4, 5, 6)

tempo_deslocamento <- c(40, 50, 20, 35, 25, 60, 15, 45, 30, 20,
                        50, 35, 25, 60, 40, 15, 45, 30, 20, 50)

# Unindo os vetores em um data frame
df <- data.frame(
  Estudante = estudante,
  Genero = genero,
  Turno = turno,
  N_disciplinas = n_disciplinas,
  Tempo_deslocamento_min = tempo_deslocamento
)

# Visualizando as primeiras linhas do data frame para conferência
head(df)




# Exercícios --------------------------------------------------------------

# 1a
#Frequência Absoluta de Gênero

# Calculando a frequência absoluta da coluna Genero
frequencia_genero <- table(df$Genero)

# Exibindo o resultado
print(frequencia_genero)


# 1b

# Calculando a frequência relativa da coluna Genero
frequencia_relativa_genero <- 100* table(df$Genero)/20

# Exibindo o resultado
cat("  M & F\n",frequencia_relativa_genero, "(%)")

# 2 -----------------------------------------------------------------------
# 2a
#Frequência Absoluta de Turno

# Calculando a frequência absoluta da coluna Turno
frequencia_turno <- table(df$Turno)

# Exibindo o resultado
print(frequencia_turno)


# 2b

# Calculando a frequência relativa da coluna turno
frequencia_relativa_turno <- 100* table(df$Turno)/20

# Exibindo o resultado
cat("  M V N\n",frequencia_relativa_turno, "(%)")



# 3 -----------------------------------------------------------------------

# a

frequencia_turno <- table(df$N_disciplinas)
print(frequencia_turno)

# b


frequencia_relativa_turno <- 100* table(df$N_disciplinas)/ 20
cat(frequencia_relativa_turno)









```
