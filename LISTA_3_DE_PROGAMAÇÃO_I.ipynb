Exercício 1

```python
num_categoria = int(input("Digite a categoria do produto (1-7): "))

# Avalia a categoria e define o custo
if num_categoria == 1:
    custo = 10
elif num_categoria == 2:
    custo = 18
elif num_categoria == 3:
    custo = 23
elif num_categoria == 4:
    custo = 26
elif num_categoria == 5:
    custo = 31
elif num_categoria == 6:
    custo = 40
elif num_categoria == 7:
    custo = 74
else:
    custo = 0 # Valor padrão caso o usuário digite uma categoria inválida

# Verifica se a categoria é válida antes de imprimir
if custo != 0:
    print(f"O produto da categoria {num_categoria} custa R$ {custo:.2f}.")
else:
    print("Categoria inválida. Por favor, digite um número entre 1 e 7.")
```
